## Imports

```python
# Python standard library
import csv

# Scipy
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt


# Project-specific
import constants
```

## Load data

```python
train_file_path = './data/train/processed/train_reformatted.csv'

df = pd.read_csv(train_file_path, sep = ',', header = 0)
```

```python
print(df.columns)
print(len(df.columns))

N_train = len(df)
N_train_iceberg = len(df[df['is_iceberg'] == 1])
N_train_ship = N_train - N_train_iceberg

print(N_train, N_train_iceberg, N_train_ship)
```

## Plot 5 random ships

```python
print(range(3, constants.num_features))
```

```python
def PlotRandomIcebergs():
    fig, axes = plt.subplots(5,2,figsize = (4,10))
    indices = np.random.randint(low = 0, high = N_train_iceberg, size = (5,))
    
    
    random_iceberg_data = df[df['is_iceberg'] == 1].iloc[indices, range(3, len(df.columns))].as_matrix()

    
    for i in range(5):
        
        
        plt.sca(axes[i%5,0])
        plt.imshow(random_iceberg_data[i,:75*75].reshape((75,75)), cmap = 'gray', origin = 'lower')
        
        plt.sca(axes[i%5,1])
        plt.imshow(random_iceberg_data[i, 75*75: 2*75*75].reshape((75,75)), cmap = 'gray', origin = 'lower')

        
    plt.show()
    
def PlotRandomShips():
    fig, axes = plt.subplots(5,2,figsize = (4,10))
    indices = np.random.randint(low = 0, high = N_train_ship, size = (5,))
    
    
    random_ship_data = df[df['is_iceberg'] == 0].iloc[indices, range(3, len(df.columns))].as_matrix()

    
    for i in range(5):
        
        
        plt.sca(axes[i%5,0])
        plt.imshow(random_ship_data[i,:75*75].reshape((75,75)), cmap = 'gray', origin = 'lower')
        
        plt.sca(axes[i%5,1])
        plt.imshow(random_ship_data[i, 75*75: 2*75*75].reshape((75,75)), cmap = 'gray', origin = 'lower')

        
    plt.show()
    
    
```

```python
PlotRandomIcebergs()
PlotRandomShips()
```
