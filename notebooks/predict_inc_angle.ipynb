## Imports

```python
# Python standard library
import csv
import json
import datetime
import sys

# Scipy
import numpy as np
import sklearn.linear_model
import sklearn.model_selection
import pandas as pd

import matplotlib.pyplot as plt

# Jupyter
import IPython.display

# Program-specific
sys.path.append('..')
import constants
import functions
```

## Load data

```python
df = functions.LoadTrainData(aug = True)
```

```python
train_df = df[df['inc_angle'] != 'na']
train_df[['inc_angle']] = train_df[['inc_angle']].apply(pd.to_numeric)
test_df = df[df['inc_angle'] == 'na']
```

```python
inputs = [input for input in constants.inputs if input != 'inc_angle']
output = ['inc_angle']
```

## Train and predict

```python
model = sklearn.linear_model.LinearRegression()


model.fit(train_df[inputs], train_df[output])
predictions = model.predict(test_df[inputs])
```

## Output

```python
ids = np.array(test_df['id'])

output_file_path = '../data/missing_inc_angles.csv'

with open(output_file_path, 'w') as file_handle:
    writer = csv.writer(file_handle, delimiter = ',')
    
    # Write header
    writer.writerow(['id', 'inc_angle'])
    
    # Write data
    for i in range(len(predictions)):
        writer.writerow([ids[i], predictions[i]])
```

## Reformat

```python
with open(output_file_path, 'r') as file_handle:
    
    lines = file_handle.readlines()
    
    edit_lines = []
    for line in lines:
        edit_lines.append(line.replace(' ','').replace('[','').replace(']','').replace('\n', ''))
    
print(edit_lines)
```
